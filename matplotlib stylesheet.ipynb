##Stylesheets

[doc page](http://matplotlib.org/users/style_sheets.html#customizing-plots-with-style-sheets)

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline

from math import pi
```

```python
x = numpy.linspace(-2*pi, 2*pi, 100)
```

```python
y = numpy.sin(x)**3
```

```python
plt.plot(x,y);
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(11,8))
plt.plot(x,y)
```

```python
print plt.style.available
```

```python
plt.style.use('bmh')
```

```python
plt.plot(x,y);
```

```python
plt.style.use(['dark_background','bmh'])
```

```python
plt.plot(x,y);
```
